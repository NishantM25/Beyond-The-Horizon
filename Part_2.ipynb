```python
import pandas as pd
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
hdulist = fits.open("lightcurve_1.fits")
```

# Task 1

```python
hdulist.info()
```

# Task 2

```python
data = hdulist[1].data
```

```python
hdulist[1].columns
```

```python
dframe = pd.DataFrame(data)
dframe
```

```python
dframe.info()
```

```python
req_data = dframe[["TIME","PDCSAP_FLUX"]]
req_data.head()
```

```python
req_data.info()
```

```python
plt.scatter(req_data["TIME"], req_data["PDCSAP_FLUX"])
plt.xlabel("Time")
plt.ylabel("PDCSAP_FLUX")
plt.title("Original")
plt.show()
```

```python
plt.scatter(req_data["TIME"], req_data["PDCSAP_FLUX"])
plt.xlabel("Time")
plt.ylabel("PDCSAP_FLUX")
plt.ylim(1e6,2.5e6)
plt.title(r"Scaling from $10^{\mathrm{6}}$ to 2.5 x $10^{\mathrm{6}}$")
plt.show()
```

```python
bonus_data = bonus_hdu[1].data
```

```python
raw_data = pd.DataFrame(bonus_data)
raw_data
```

```python
plt.scatter(raw_data["XI"], raw_data["ETA"], s=5, c='b', marker='o')
```

```python
plt.scatter(raw_data["XI_CORR"], raw_data["ETA_CORR"], c='r',s=5, marker='o')
```

```python
plt.imshow(raw_data[["XI", "ETA"]],  cmap='viridis', aspect='auto')
plt.show()
```

```python
plt.imshow(raw_data[["XI_CORR", "ETA_CORR"]],  cmap='magma', aspect='auto')
plt.show()
```

```python
plt.pcolormesh(raw_data[["XI_CORR", "ETA_CORR"]], cmap='plasma')
```
