```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE
from sklearn import linear_model
```

```python
raw_data = pd.read_csv("labelled_data.csv")
raw_data.head()
```

```python
X = raw_data.drop("LABEL", axis=1)
y = raw_data["LABEL"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
smote = SMOTE(sampling_strategy='auto', random_state=36)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
linear_classifier = linear_model.LinearRegression()
linear_classifier.fit(X_train_resampled, y_train_resampled)
```

```python
y_pred = linear_classifier.predict(X_test)
predicted_labels = np.where(y_pred > 1.4, 2, 1)
```

```python
accuracy = accuracy_score(y_test, predicted_labels)
print(accuracy)
```
